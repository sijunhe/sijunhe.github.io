```python
prev_1, recall_1, support_1 = 0.5, 0.2, 50
prev_2, recall_2, support_2 = 0.8, 0.5, 150
f_1 = 2 * (prev_1 * recall_1) / (prev_1 + recall_1)
f_2 = 2 * (prev_2 * recall_2) / (prev_2 + recall_2)
print((f_1 * support_1 + f_2 * support_2) / (support_1 + support_2))
```

```python
weighted_prev = (prev_1 * support_1 + prev_2 * support_2) / (support_1 + support_2)
weighted_recall = (recall_1 * support_1 + recall_2 * support_2) / (support_1 + support_2)
2 * weighted_prev * weighted_recall / (weighted_recall + weighted_prev)
```
